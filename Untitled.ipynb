# Thresholded Image

```python
# use functions defined in the lessons for threshold

#sobel: "derivative"
def abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255)):
    # Calculate directional gradient
    # Apply threshold
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    if orient == 'x':
        sobel = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize = sobel_kernel)
    if orient =='y':
        sobel = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize = sobel_kernel)
    abso = np.absolute(sobel)
    scaled = np.uint8(255*abso/np.max(abso))
    
    binary_output = np.zeros_like(scaled) 
    binary_output[(scaled>thresh[0]) &(scaled<thresh[1])]=1
    return binary_output

#magnitude threshold: defines a threshold for the magnitude of the gradient
def mag_thresh(image, sobel_kernel=3, mag_thresh=(0, 255)):
    # Calculate gradient magnitude
    # Apply threshold
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=sobel_kernel)
    sobely = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=sobel_kernel)
    magn = np.sqrt(sobelx**2 + sobely**2)
    scaled = np.uint8(255*magn/np.max(magn))
    binary_output = np.zeros_like(scaled) 
    binary_output[(scaled>mag_thresh[0]) &(scaled<mag_thresh[1])]=1
    return binary_output
    

# direction threshold: only use pixels where the gradient points into the directions within the threshold
def dir_threshold(image, sobel_kernel=3, thresh=(0, np.pi/2)):
    # Calculate gradient direction
    # Apply threshold
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    sobelx = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=sobel_kernel)
    sobely = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=sobel_kernel)
    absx = np.absolute(sobelx)
    absy = np.absolute(sobely)
    direct = np.arctan2(absy,absx)
    binary_output = np.zeros_like(direct) 
    binary_output[(direct>thresh[0]) &(direct<thresh[1])]=1
    return binary_output

# color thresholds
def hls_select(img, thresh=(0, 255),thresh2=(0,255),thresh3=(0,255)):
    
    hls = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
    s = hls[:,:,2]
    l = hls[:,:,1]
    h = hls[:,:,0]
    binary_output = np.zeros_like(s)
    binary_output[ (s>thresh[0]) & (s<=thresh[1])&(l>thresh2[0])&(l<thresh2[1])&(s>thresh3[0])&(s<thresh3[1])]=1
    return binary_output
```

```python
def combine(image,ksize=7):
    #combine thresholds
    gradx = abs_sobel_thresh(image, orient='x', sobel_kernel=ksize, thresh=(60, 100))
    grady = abs_sobel_thresh(image, orient='y', sobel_kernel=ksize, thresh=(60, 100))
    mag_binary = mag_thresh(image, sobel_kernel=ksize, mag_thresh=(70, 255))
    dir_binary = dir_threshold(image, sobel_kernel=ksize, thresh=(np.pi/5, 5*np.pi/14))
    hls = hls_select(image,(100,255),(0,200))

    combined = np.zeros_like(dir_binary)
    combined[(dir_binary==1)&(((gradx == 1) & (grady == 1)) | ((mag_binary == 1) | (hls==1)))] = 1
    return combined
image = cv2.imread("test_images/test3.jpg")
combined = combine(image)
mpimg.imsave('output_images/test3_binary.jpg',combined)
#plot 
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 9))
f.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=50)
ax2.imshow(combined, cmap='gray')
ax2.set_title('Thresholded Gradient', fontsize=50)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```

# Perspective Transform


```python
# use straight line picture to determine corner points for the perspective transform
image = cv2.imread('test_images/straight_lines1.jpg')
pts = np.array([[562,470],[715,470],[1090,720],[200,720]], np.int32)

pts = pts.reshape((-1,1,2))
image = cv2.polylines(image, [pts], True, (0,255,0), 3)

plt.imshow(image)
```

```python
def undistort(img,mtx,dist):
    return cv2.undistort(img,mtx,dist,None,mtx)

#as in the lessons define a function to calculate the perspective transform matrix
def persp_transf(undist,pts):
    gray = undist
    offset1 = 300
    offset2 = 0
    dst = np.float32([[offset1,offset2],[gray.shape[1]-offset1,offset2],[gray.shape[1]-offset1,gray.shape[0]],[offset1,gray.shape[0]]])
    src = np.float32(pts)
    M = cv2.getPerspectiveTransform(src, dst)
    return M

def unwarp(M,undist):
    warped = cv2.warpPerspective(undist, M, (undist.shape[1],undist.shape[0]), flags=cv2.INTER_LINEAR)
    return warped
#all together:
def transform(img,mtx,dist,pts):
    undist = undistort(img,mtx,dist)
    M = persp_transf(undist,pts)
    warped = unwarp(M,undist)
    return warped,M
```

# Lane Finding

```python
#now use functions to find lane lines and the corresponding polynomial
def find_lane_pixels(binary_warped,leftx_base = None,rightx_base = None):
    if leftx_base == None or rightx_base == None:
        # Take a histogram of the bottom half of the image
        histogram = np.sum(binary_warped[binary_warped.shape[0]//2:,:], axis=0)
        # Create an output image to draw on and visualize the result
        out_img = np.dstack((binary_warped, binary_warped, binary_warped))
        # Find the peak of the left and right halves of the histogram
        # These will be the starting point for the left and right lines
        midpoint = np.int(histogram.shape[0]//2)
        leftx_base = np.argmax(histogram[:midpoint])
        rightx_base = np.argmax(histogram[midpoint:]) + midpoint

    # HYPERPARAMETERS
    # Choose the number of sliding windows
    nwindows = 9
    # Set the width of the windows +/- margin
    margin = 100
    # Set minimum number of pixels found to recenter window
    minpix = 30

    # Set height of windows - based on nwindows above and image shape
    window_height = np.int(binary_warped.shape[0]//nwindows)
    # Identify the x and y positions of all nonzero pixels in the image
    nonzero = binary_warped.nonzero()
    nonzeroy = np.array(nonzero[0])
    nonzerox = np.array(nonzero[1])
    # Current positions to be updated later for each window in nwindows
    leftx_current = leftx_base
    rightx_current = rightx_base

    # Create empty lists to receive left and right lane pixel indices
    left_lane_inds = []
    right_lane_inds = []

    # Step through the windows one by one
    for window in range(nwindows):
        # Identify window boundaries in x and y (and right and left)
        win_y_low = binary_warped.shape[0] - (window+1)*window_height
        win_y_high = binary_warped.shape[0] - window*window_height
        
        win_xleft_low = leftx_current - margin  
        win_xleft_high = leftx_current + margin  
        win_xright_low = rightx_current - margin  
        win_xright_high = rightx_current + margin  
        
        # Draw the windows on the visualization image
        cv2.rectangle(out_img,(win_xleft_low,win_y_low),
        (win_xleft_high,win_y_high),(0,255,0), 2) 
        cv2.rectangle(out_img,(win_xright_low,win_y_low),
        (win_xright_high,win_y_high),(0,255,0), 2) 
        
        ### TO-DO: Identify the nonzero pixels in x and y within the window ###
        good_left_inds = ((nonzerox >=win_xleft_low )&(nonzerox<win_xleft_high)&
                            (nonzeroy >=win_y_low)&(nonzeroy<win_y_high)).nonzero()[0]
        good_right_inds = ((nonzerox >=win_xright_low )&(nonzerox<win_xright_high)&
                            (nonzeroy >=win_y_low)&(nonzeroy<win_y_high)).nonzero()[0]
        # Append these indices to the lists
        left_lane_inds.append(good_left_inds)
        right_lane_inds.append(good_right_inds)
        
        if len(good_left_inds)>minpix:
            leftx_current = np.int(np.mean(nonzerox[good_left_inds]))
        if len(good_right_inds)>minpix:
            rightx_current = np.int(np.mean(nonzerox[good_right_inds]))   
        # Remove this when you add your function

    # Concatenate the arrays of indices (previously was a list of lists of pixels)
    
    left_lane_inds = np.concatenate(left_lane_inds)
    right_lane_inds = np.concatenate(right_lane_inds)

    # Extract left and right line pixel positions
    leftx = nonzerox[left_lane_inds]
    lefty = nonzeroy[left_lane_inds] 
    rightx = nonzerox[right_lane_inds]
    righty = nonzeroy[right_lane_inds]

    return leftx, lefty, rightx, righty, out_img
leftx,lefty,rightx,righty ,out_img = find_lane_pixels(trans)
plt.imshow(out_img)
```

```python
def fit_polynomial(binary_warped,image,M, leftx_base = None,rightx_base = None):
    # Find our lane pixels first
    leftx, lefty, rightx, righty, out_img = find_lane_pixels(binary_warped,leftx_base,rightx_base)

    ###  Fit a second order polynomial to each using `np.polyfit` ###
    left_fit = np.polyfit(lefty,leftx,2)
    right_fit = np.polyfit(righty,rightx,2)
    
    # Generate x and y values for plotting
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]

    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    y_eval = np.max(ploty)
    
    left_fitm = np.polyfit(lefty*ym_per_pix,leftx*xm_per_pix,2)
    left_curverad = (1+(2*left_fitm[0]*y_eval*ym_per_pix + left_fitm[1])**2)**(1.5)/(2*np.abs(left_fitm[0]))  
    right_fitm = np.polyfit(righty*ym_per_pix,rightx*xm_per_pix,2)
    right_curverad = (1+(2*right_fitm[0]*y_eval*ym_per_pix + right_fitm[1])**2)**(1.5)/(2*np.abs(right_fitm[0]))  
  
    right_linex=right_fit[0]*y_eval**2 + right_fit[1]*y_eval + right_fit[2]
    left_linex=left_fit[0]*y_eval**2 + left_fit[1]*y_eval + left_fit[2]
    middle = (right_linex+left_linex)/2 
    

    
    position = (binary_warped.shape[1]//2 - middle)*xm_per_pix# I assume that the position of the vehicle is the center of the picture

    
    left_line = (left_fit[0]*binary_warped.shape[0]**2 + left_fit[1]*binary_warped.shape[0]+left_fit[2])*xm_per_pix
    right_line = (right_fit[0]*binary_warped.shape[0]**2 + right_fit[1]*binary_warped.shape[0]+right_fit[2])*xm_per_pix
   
    return left_curverad,right_curverad,position, left_fit,right_fit,left_line,right_line,leftx,lefty,rightx,righty
```

# Pipeline

```python
# define a mask to apply for the next image, when lane lines are detected
def area_mult(transf,lfit,rfit,margin=50):
    mask = np.zeros_like(transf)
    for x in range(transf.shape[1]):
        for y in range(transf.shape[0]):
            value_for_lfit = lfit[0]*y**2 + lfit[1]*y +lfit[2]
            value_for_rfit = rfit[0]*y**2 + rfit[1]*y +rfit[2]
            if x>value_for_lfit-margin and x < value_lfor_fit + margin:
                mask[y][x]=1
            if x>value_for_rfit-margin and x < value_for_rfit + margin:
                mask[y][x]=1
    masked_image = transf*mask
    return masked_image
```

```python
#draw lane lines
def draw(binary_warped,left_fit,right_fit,M,image):
    
    ploty = np.linspace(0, binary_warped.shape[0]-1, binary_warped.shape[0] )
    
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]   
    
    # Create an image to draw the lines on
    warp_zero = np.zeros_like(binary_warped).astype(np.uint8)
    color_warp = np.dstack((warp_zero, warp_zero, warp_zero))

    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))
    Minv = np.linalg.inv(M)
    # Warp the blank back to original image space using inverse perspective matrix (Minv)
    newwarp = cv2.warpPerspective(color_warp, Minv, (image.shape[1], image.shape[0])) 
    # Combine the result with the original image
    result = cv2.addWeighted(image, 1, newwarp, 0.3, 0)
    plt.imshow(result)
    return result
```

```python
#all together
def helper(image,leftx_base=None,rightx_base = None,discovered = False,left_fit = None,right_fit = None):
    undist = cv2.undistort(image,mtx,dist,None,mtx)
    combined = combine(undist)
    transf,M = transform(combined,mtx,dist,pts)
    if discovered:# if in the image before lines were discovered use these to determine the area where the new line can be
        transf = area_mult(transf,left_fit,right_fit,10)
    left,right,pos,lfit,rfit,lline,rline,leftx,lefty,rightx,righty = fit_polynomial(transf,image,M,leftx_base,rightx_base)
    return left,right,pos,lfit,rfit,lline,rline,leftx,lefty,rightx,righty,transf,M

def pipeline(image):
    left,right,pos,lfit,rfit,lline,rline,leftx,lefty,rightx,righty,transf,M = helper(image)
    res = draw(transf,lfit,rfit,M,image)
    return res
#define another function that draws the lane lines, given values for lfit,rfit,M (for the case when the currently 
#detected lane lines are not good)
def pipel(image,lfit,rfit,M,transf):
    res = draw(transf,lfit,rfit,M,image)
    return res
```

```python
class Line():
    def __init__(self):
        # number of past values to be saved
        self.n_values = 10
        # was the line detected in the last iteration?
        self.detected = False  
        # x values of the last n fits of the line
        self.recent_xfitted = [] 
        #average x values of the fitted line over the last n iterations
        self.bestx = None     
        #polynomial coefficients averaged over the last n iterations
        self.best_fit = None  
        #polynomial coefficients for the most recent fit
        self.current_fit = [np.array([False])]  
        #recent values for the polynomial coefficients
        self.recent_fit = []
        #radius of curvature of the line in some units
        self.radius_of_curvature = None 
        #distance in meters of vehicle center from the line
        self.line_base_pos = None 
        #difference in fit coefficients between last and new fits
        self.diffs = np.array([0,0,0], dtype='float') 
        #x values for detected line pixels
        self.allx = None  
        #y values for detected line pixels
        self.ally = None  
```

```python
right_line = Line()
left_line = Line()

#write a function to determine all values for the lines detected in an image and decide wheter these values are reasonable
def process_lines(image,left_line,right_line):
    #new lines
    new_left_line = Line()
    new_right_line =  Line()
    #if the old lines were detected correctly use the helper function which only searches near the old lines
    if left_line.detected == True and right_line.detected == True:
        new_left_line.radius_of_curvature,new_right_line.radius_of_curvature,pos,\
        new_left_line.current_fit,new_right_line.current_fit,lline,rline,new_left_line.allx,new_left_line.ally,\
        new_right_line.allx,new_right_line.ally,tranf,M = helper(image,left_line.bestx,right_line.bestx,True,left_line.current_fit,right_line.current_fit)
    else:
        new_left_line.radius_of_curvature,new_right_line.radius_of_curvature,pos,\
        new_left_line.current_fit,new_right_line.current_fit,lline,rline,new_left_line.allx,new_left_line.ally,\
        new_right_line.allx,new_right_line.ally,tranf,M = helper(image)
    
    new_left_line.recent_fit = left_line.recent_fit
    new_left_line.recent_fit.append(new_left_line.current_fit)
    if len(new_left_line.recent_fit)>5:
        new_left_line.recent_fit.pop(0)
    new_right_line.recent_fit = right_line.recent_fit
    new_right_line.recent_fit.append(new_right_line.current_fit)
    if len(new_right_line.recent_fit)>5:
        new_right_line.recent_fit.pop(0)    
    
    
    
    new_left_line.best_fit = sum(new_left_line.recent_fit)/len(new_left_line.recent_fit)
    new_right_line.best_fit = sum(new_right_line.recent_fit)/len(new_right_line.recent_fit)
    
    new_right_line.recent_xfitted = right_line.recent_xfitted
    new_right_line.recent_xfitted.append(rline)
    if len(new_right_line.recent_xfitted)>right_line.n_values:
        new_right_line.recent_xfitted.pop(0)
    new_left_line.recent_xfitted = left_line.recent_xfitted
    new_left_line.recent_xfitted.append(lline)
    if len(new_left_line.recent_xfitted)>left_line.n_values:
        left_line.recent_xfitted.pop(0)
    
   
    
    new_right_line.bestx = sum(new_right_line.recent_xfitted)/len(new_right_line.recent_xfitted)
    new_left_line.bestx = sum(new_left_line.recent_xfitted)/len(new_left_line.recent_xfitted)
    
    new_right_line.diffs = new_right_line.current_fit-right_line.current_fit
    new_left_line.diffs  = new_left_line.current_fit - left_line.current_fit
    
    new_right_line.line_base_pos = new_right_line.bestx - pos
    new_left_line.line_base_pos = pos - new_left_line.bestx
    
    # now, having all the data, check if the values are reasonable
    good = True #bool to determine if the values are reasonable
    #width of the lane
    lane_lines_distance = new_right_line.line_base_pos + new_left_line.line_base_pos
    
    if lane_lines_distance <3.5 or lane_lines_distance >4.2: #minimum lane width: 3.7m, so take a little less as threshold
        good = False
    # now check if the width at the upper end is also within a certain range
    if (new_right_line.current_fit[2]-new_left_line.current_fit[2])*3.7/700 < 3.3:
        good = False
    if (new_right_line.current_fit[2]-new_left_line.current_fit[2])*3.7/700 > 4.5:
        good = False
    
    #if the values are reasonable return the new detected lines
    if good==True:
        return new_left_line,new_right_line,good
    #else use the old lines
    else:
        return left_line,right_line,good

#function to use in fl_image(), that processes every image    
def process_image(image):
    global left_line 
    global right_line
    
    left_line,right_line,good = process_lines(image,left_line,right_line)

    if  good==True:
        out = pipeline(image)
    else:
        # in case of not well detected lines use helper (doesn't draw the lines) to determine transf and M
        left,right,pos,lfit,rfit,lline,rline,leftx,lefty,rightx,righty,transf,M = helper(image)
        # and then draw the old lines
        out = pipel(image,left_line.current_fit,right_line.current_fit,M,transf)
    return out
```
